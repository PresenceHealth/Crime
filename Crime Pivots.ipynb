```python
import numpy as np
import pandas as pd
```

```python
crime_data = pd.read_csv("Chicago_Crimes.csv") #h
location = pd.read_csv("Chicago Crimes geocoded to ward cca tract.csv",dtype={'tract: FID': str}) #v
```

```python
crime_data['year'] = crime_data.date.str.replace(pat='-\d\d-\d\d \d\d:\d\d:\d\d', repl='')
```

```python
EXTRA = ['community_area', 'district', 'id', 'latitude', 'longitude', 'ward', 'ID', 'Longitude', 'Latitude', 'ward: FID', 'cca: FID','tract: FID']

def chicago_pivot(crime_data, location,geog, drop_col=EXTRA):
    d = pd.merge(crime_data, location, left_on='id',right_on='ID')
    drop_col = [x for x in drop_col if x != geog]
    d.drop(drop_col, axis=1, inplace=True)
    out = d.pivot_table(index=geog, columns=['year'], aggfunc=sum)
    out.columns = ['_'.join(col).strip() for col in out.columns.values]
    out.index.name = None
    return out

```

```python
chicago_pivot(crime_data, location, 'tract: FID').to_csv('chicago_crime_by_tract.csv')
```

```python
ward = chicago_pivot(crime_data, location, 'ward: FID').to_csv('chicago_crime_by_ward.csv')
cca = chicago_pivot(crime_data, location, 'cca: FID').to_csv('chicago_crime_by_cca.csv')
```
